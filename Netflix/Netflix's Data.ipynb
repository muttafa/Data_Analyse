```python
sns.countplot(x = data["type"] , data=data)
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.title("Tür")
```

```python
sns.countplot(x = data["rating"] , data=data)
sns.countplot(x = data["rating"] , data=data)
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title("Ratings")
```

```python
plt.figure(figsize = (10,8))
sns.countplot(x="rating",hue="type", data = data)
plt.title("Tür ve Rating Oranları")
plt.show()
```

```python
labels = ["Movies","TV Shows"]
size = data["type"].value_counts()
colors = plt.cm.Wistia(np.linspace(0,1,2))
explode = [0,0.1]
plt.rcParams["figure.figsize"] = (9,9)
plt.pie(size,labels=labels,colors=colors,explode=explode,shadow=True,startangle = 90)
plt.title("Tür Görünümü",fontsize = 25)
plt.legend()
plt.show()
```

```python
data["rating"].value_counts().plot.pie(autopct="%1.1f%%",shadow=True,figsize=(10,8))
plt.show()
```
